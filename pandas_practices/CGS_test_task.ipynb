**Task 9 (1.5 point).** List top-5 tuples of most correlated **player's characteristics** in the descending order of the absolute [Pearson's coefficient](https://en.wikipedia.org/wiki/Pearson_correlation_coefficient) value.

**Note 1:** Players characteristics are all the columns in `Player_Attributes` table except `[id, player_fifa_api_id, player_api_id, date, preferred_foot, attacking_work_rate, defensive_work_rate]`). <br>
**Note 2:** Exclude duplicated pairs from the list. E.g. ('gk_handling', 'gk_reflexes') and ('gk_reflexes', 'gk_handling') are duplicates, leave just one of them in the resulting list.

```python
player_attributes = pd.read_csv('player_attributes.csv', index_col="Unnamed: 0")
```

```python
player_attributes.head()
```

```python
import numpy as np

excluded_columns = ['id', 'player_fifa_api_id', 'player_api_id', 'date', 'preferred_foot', 'attacking_work_rate', 'defensive_work_rate']
player_characteristics = player_attributes.drop(columns=excluded_columns)
# Calculate the correlation matrix
correlation_matrix = player_characteristics.corr()
# Extract the upper triangle of the correlation matrix (excluding the diagonal)
upper_triangle = correlation_matrix.where(np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool))
# Get the top-5 correlated pairs in descending order of absolute correlation
top_correlations = (upper_triangle.abs().unstack().sort_values(ascending=False).drop_duplicates().head(5))
# Convert the top correlations into a list of tuples
top_correlated_features = [tuple(index) for index, value in top_correlations.items()]
```

**Task 14 (2 points).** Find a team with the highest win rate in the **2015/2016** season. Win rate means won matches / all matches. If there are several teams with the highest win rate return the first by name in lexical order

```python
# Filter matches for the 2015/2016 season
matches_2015_2016 = match_data[match_data['season'] == '2015/2016']

# Calculate the total number of matches played by each team
matches_played_home = matches_2015_2016['home_team_api_id'].value_counts()
matches_played_away = matches_2015_2016['away_team_api_id'].value_counts()
matches_played_total = matches_played_home.add(matches_played_away, fill_value=0)

# Calculate the number of matches won by each team
home_wins = matches_2015_2016[matches_2015_2016['home_team_goal'] > matches_2015_2016['away_team_goal']]['home_team_api_id']
away_wins = matches_2015_2016[matches_2015_2016['home_team_goal'] < matches_2015_2016['away_team_goal']]['away_team_api_id']
total_wins = pd.concat([home_wins, away_wins], ignore_index=True)
wins_count = total_wins.value_counts()

# Calculate the win rate for each team
win_rate = wins_count / matches_played_total

# Find the team with the highest win rate
team_with_highest_win_rate = win_rate.idxmax()

# Retrieve the team name from team_data
team_name = team_data.loc[team_data['team_api_id'] == team_with_highest_win_rate, 'team_long_name'].iloc[0]

team_highest_winrate_2015_2016 = team_name
```
